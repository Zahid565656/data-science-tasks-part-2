Import Libraries

```python
import pandas as pd
import plotly.express as px
```

Load Dataset

```python
df = pd.read_excel("Global_Superstore.xlsx")
```

Explore Dataset

```python
df.head()
df.info()
df.shape
df.isnull().sum()
```

Calculate KPIs


Total Sales

```python
total_sales = df['Sales'].sum()
print("Total Sales:", total_sales)
```

Total Profit

```python
total_profit = df['Profit'].sum()
print("Total Profit:", total_profit)
```

Top 5 Customers

```python
top_customers = (
    df.groupby('Customer Name')['Sales']
    .sum()
    .sort_values(ascending=False)
    .head(5)
)

top_customers
```

Sales by Category

```python
sales_category = (
    df.groupby('Category')['Sales']
    .sum()
    .reset_index()
)

fig = px.bar(
    sales_category,
    x='Category',
    y='Sales',
    title='Sales by Category'
)

fig.show()
```

Profit by Region

```python
profit_region = (
    df.groupby('Region')['Profit']
    .sum()
    .reset_index()
)

fig = px.bar(
    profit_region,
    x='Region',
    y='Profit',
    title='Profit by Region'
)

fig.show()
```

Top 5 Customers Chart

```python
top_df = top_customers.reset_index()

fig = px.bar(
    top_df,
    x='Customer Name',
    y='Sales',
    title='Top 5 Customers by Sales'
)

fig.show()
```

An interactive dashboard was developed using Streamlit and Plotly to analyze Global Superstore data. The dashboard provides KPIs such as total sales and profit and allows users to explore sales performance across regions, categories, and top customers through interactive visualizations.
